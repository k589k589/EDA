```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import plotly.express as px
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
ALLsurvey=pd.read_csv('../input/kaggle-survey-2021/kaggle_survey_2021_responses.csv')
ALLsurvey.drop(0,inplace=True)
ALLsurvey.head()
```

```python
import warnings
warnings.filterwarnings("ignore")
if not os.path.exists('/kaggle/working/individual_charts/'):
    os.mkdir('/kaggle/working/individual_charts/')

def load_csv(base_dir,file_name):
    file_path = os.path.join(base_dir,file_name)
    df = pd.read_csv(file_path,low_memory=False,encoding='ISO-8859-1')
    return df    

base_dir_2017 = '/kaggle/input/kaggle-survey-2017/'
file_name_2017 = 'multipleChoiceResponses.csv'
survey_df_2017 = load_csv(base_dir_2017,file_name_2017)
survey_df_2017 = survey_df_2017[survey_df_2017['Country']=='Taiwan'] 
responses_df_2017 = survey_df_2017[1:]
survey_df_2017.to_csv('2017_kaggle_ds_and_ml_survey_data_scientists_in_india_responses_only.csv',index=False)


base_dir_2019 = '/kaggle/input/kaggle-survey-2019/'
file_name_2019 = 'multiple_choice_responses.csv'
survey_df_2019 = load_csv(base_dir_2019,file_name_2019)
survey_df_2019 = survey_df_2019[survey_df_2019['Q3']=='Taiwan']
responses_df_2019 = survey_df_2019[1:]
survey_df_2019.to_csv('2019_kaggle_ds_and_ml_survey_data_scientists_in_india_responses_only.csv',index=False)

base_dir_2020 = '/kaggle/input/kaggle-survey-2020'
file_name_2020 = 'kaggle_survey_2020_responses.csv'
survey_df_2020 = load_csv(base_dir_2020,file_name_2020)
survey_df_2020 = survey_df_2020[survey_df_2020['Q3']=='Taiwan'] 
responses_df_2020 = survey_df_2020[1:]
survey_df_2020.to_csv('2020_kaggle_ds_and_ml_survey_data_scientists_in_india_responses_only.csv',index=False)

base_dir_2021 = '../input/kaggle-survey-2021/'
file_name_2021 = 'kaggle_survey_2021_responses.csv'
survey_df_2021 = load_csv(base_dir_2021,file_name_2021)
survey_df_2021 = survey_df_2021[survey_df_2021['Q3']=='Taiwan'] 
responses_df_2021 = survey_df_2021[1:]
survey_df_2021.to_csv('2021_kaggle_ds_and_ml_survey_data_scientists_in_india_responses_only.csv',index=False)
```

# Company ML situation
* 我們可以發現有部署機器模型的公司是佔10.7與12.2，差距並不大
* 在使用機器學習產生見解方面就有明顯差異，分別為21.3%與16.8%，台灣明顯有了落後，我認為很多產業不一定需要部署機器學習模型，但是藉由機器學習影響公司營運方向是未來趨勢，也就是我們說的Data-Driven。
* We can find that companies that have deployed machine models account for 10.7 and 12.2, and the gap is not large
* There is a significant difference in the use of machine learning to generate insights, 21.3% and 16.8% respectively. Taiwan is clearly lagging behind. I think that many industries do not necessarily need to deploy machine learning models, but it is the future to use machine learning to influence company operations.Which is what we call Data-Driven

```python
columns=['Q40_Part_1','Q40_Part_2','Q40_Part_3','Q40_Part_4','Q40_Part_5','Q40_Part_6','Q40_Part_7',
        'Q40_Part_8','Q40_Part_9','Q40_Part_10','Q40_Part_11','Q40_OTHER']

Q40=ALLsurvey[columns]
x=pd.DataFrame()
for i in columns:
    y=Q40[i].value_counts()
    x=x.append(y)

Q40=x.stack().reset_index().iloc[:,1:]
recols={'level_1':'Course',0:'Counts'}
Q40=Q40.rename(columns=(recols))
Q40=Q40.set_index('Course')
```

```python
plt.figure(figsize=(10,10))
plt.title('(ALL WORLD)On which platforms have you begun or completed data science courses?',fontsize=20)
sns.barplot(y=Q40.index,x=Q40['Counts'],palette='Greens_r',order=Q40['Counts'].sort_values(ascending=False).index)
```

```python
TWQ40=responses_df_2021[columns]

x=pd.DataFrame()
for i in columns:
    y=TWQ40[i].value_counts()
    x=x.append(y)

TWQ40=x.stack().reset_index().iloc[:,1:]
recols={'level_1':'Course',0:'Counts'}
TWQ40=TWQ40.rename(columns=(recols))
TWQ40=TWQ40.set_index('Course')

plt.figure(figsize=(10,10))
plt.title('(In Taiwan),On which platforms have you begun or completed data science courses?')
sns.barplot(y=TWQ40.index,x=TWQ40['Counts'],palette='Greens_r',order=TWQ40['Counts'].sort_values(ascending=False).index)
```

```python
learn=pd.DataFrame()
learn['Course']=Q40.index
learn['Taiwan']=(TWQ40['Counts']/333).values
learn['AllWorld']=(Q40['Counts']/25973).values
px.line(learn,x='Course',y=['Taiwan','AllWorld'],title='Percentage of people learning from which platform')
```
